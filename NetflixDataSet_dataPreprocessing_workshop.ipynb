6. Adding new features

We can add new features from the date_added column which are month and year.

```python
df.shape
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],format='mixed')# convert the string dates to date time format
df['year_added']=df['date_added'].dt.year#extract the year from date_added and create new colum added_year and insert into it the value that we extracted
df['month_added']=df["date_added"].dt.month
df.head()
```

```python
df['rating'].unique()
```

```python
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
df['rating'].replace(ratings_ages,inplace=True)#we replaced the old ratings values by age rate corresponding to the previous ratings through the dictionary that we created
df.head()
```
